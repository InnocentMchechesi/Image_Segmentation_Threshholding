```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# 1. Load the image
img = cv2.imread('my_img.jpeg')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# 2. Pre-processing: Extracting the Red channel for maximum contrast against blue
# This helps separate the white and black objects from the blue table
gray = img[:,:,2] 

# Apply a slight Gaussian Blur to reduce noise before thresholding
blurred = cv2.GaussianBlur(gray, (5, 5), 0)

# 3. Global Thresholding (Fixed Value)
# We pick a manual threshold value (e.g., 127)
ret, thresh_global = cv2.threshold(blurred, 127, 255, cv2.THRESH_BINARY)

# 4. Otsu's Binarization
# This automatically calculates the best threshold value based on the histogram
ret_otsu, thresh_otsu = cv2.threshold(blurred, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

# 5. Adaptive Thresholding
# This calculates thresholds for small regions, handling lighting variations better
thresh_adaptive = cv2.adaptiveThreshold(blurred, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, 
                                        cv2.THRESH_BINARY, 11, 2)

# 6. Cleaning up the results (Morphological Operations)
kernel = np.ones((5,5), np.uint8)
closing = cv2.morphologyEx(thresh_otsu, cv2.MORPH_CLOSE, kernel)

# 7. Visualization
titles = ['Original Image', 'Grayscale (Red Channel)', 'Global (v=127)', 
          "Otsu's Method", 'Adaptive Gaussian', 'Cleaned (Otsu + Closing)']

images = [img_rgb, gray, thresh_global, thresh_otsu, thresh_adaptive, closing]

plt.figure(figsize=(18, 10))
for i in range(6):
    plt.subplot(2, 3, i+1)
    plt.imshow(images[i], 'gray' if i > 0 else None)
    plt.title(titles[i])
    plt.xticks([]), plt.yticks([])

plt.tight_layout()
plt.show()

# 8. Object Extraction (Example using the Cleaned Mask)
# We invert the mask if necessary so the objects are white (255)
mask = cv2.bitwise_not(closing) 
extracted = cv2.bitwise_and(img_rgb, img_rgb, mask=mask)

plt.figure(figsize=(10, 5))
plt.imshow(extracted)
plt.title('Extracted Objects')
plt.axis('off')
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def show(title, img, cmap='gray'):
    plt.figure(figsize=(6,6))
    plt.title(title)
    plt.imshow(img, cmap=cmap)
    plt.axis('off')

img = cv2.imread('my_img.jpeg')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

show("Original Image", img_rgb)


# 1. Global Thresholding
_, th_global = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)


# 2. Otsu Thresholding
_, th_otsu = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)


# 3. Adaptive Mean Thresholding
th_adapt_mean = cv2.adaptiveThreshold(
    gray, 255,
    cv2.ADAPTIVE_THRESH_MEAN_C,
    cv2.THRESH_BINARY,
    11, 2
)


# 4. Adaptive Gaussian Thresholding
th_adapt_gauss = cv2.adaptiveThreshold(
    gray, 255,
    cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
    cv2.THRESH_BINARY,
    11, 2
)

show("Global Threshold", th_global)
show("Otsu Threshold", th_otsu)
show("Adaptive Mean", th_adapt_mean)
show("Adaptive Gaussian", th_adapt_gauss)


# Function to draw bounding boxes
def draw_boxes(binary, original, title):
    img_copy = original.copy()
    
    contours, _ = cv2.findContours(binary, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    
    for i, cnt in enumerate(contours):
        area = cv2.contourArea(cnt)
        
        if area > 2000:
            x, y, w, h = cv2.boundingRect(cnt)
            cv2.rectangle(img_copy, (x,y), (x+w,y+h), (0,255,0), 2)
            cv2.putText(img_copy, f'Obj{i+1}', (x, y-10),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.6, (255,0,0), 2)
    
    show(title, img_copy, cmap=None)


# Draw results
draw_boxes(th_global, img_rgb, "Global Threshold Detection")
draw_boxes(th_otsu, img_rgb, "Otsu Threshold Detection")
draw_boxes(th_adapt_mean, img_rgb, "Adaptive Mean Detection")
draw_boxes(th_adapt_gauss, img_rgb, "Adaptive Gaussian Detection")
```
